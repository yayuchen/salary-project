```python
def plot_feature(df, col):
    '''
    Make plot for each features
    left, the distribution of samples on the feature
    right, the dependance of salary on the feature
    '''
    plt.figure(figsize = (14, 6))
    plt.subplot(1, 2, 1)
    if df[col].dtype == 'int64':
        df[col].value_counts().sort_index().plot()
    else:
        #change the categorical variable to category type and order their level by the mean salary
        #in each category
        mean = df.groupby(col)['salary'].mean()
        df[col] = df[col].astype('category')
        levels = mean.sort_values().index.tolist()
        df[col].cat.reorder_categories(levels, inplace=True)
        df[col].value_counts().plot()
    plt.xticks(rotation=45)
    plt.xlabel(col)
    plt.ylabel('Counts')
    plt.subplot(1, 2, 2)

    if df[col].dtype == 'int64' or col == 'companyId':
        #plot the mean salary for each category and fill between the (mean - std, mean + std)
        mean = df.groupby(col)['salary'].mean()
        std = df.groupby(col)['salary'].std()
        mean.plot()
        plt.fill_between(range(len(std.index)), mean.values-std.values, mean.values + std.values, \
                         alpha = 0.1)
    else:
        sns.boxplot(x = col, y = 'salary', data=df)
    
    plt.xticks(rotation=45)
    plt.ylabel('Salaries')
    plt.show()
```

```python
def transform_label(df, col):
    cat_dict = {}
    cats = df[col].cat.categories.tolist()
    for cat in cats:
        cat_dict[cat] = df[df[col]==cat]['salary'].mean()
    df[col] = df[col].map(cat_dict)
```

```python
for i in df.columns:
    if df[i].dtype.name == 'category':
        transform_label(df, i)
```

```python
# input_features_above_3 = ['jobType', 'degree', 'major', 'industry', 'yearsExperience', 'milesFromMetropolis']
from sklearn.preprocessing import LabelEncoder

def Label_encoder(df, col):
    le = LabelEncoder()
    le.fit(df[col])
    df[col] = le.transform(df[col])
```

```python
encoded_df = full_df.copy()
for i in input_features_above_3:
    Label_encoder(encoded_df, i)    
```

```python
def groupby_feature(df, col):
    groups= df.groupby(by=col)
    group_col_stat = pd.DataFrame({'group_mean': groups['salary'].mean()})
    group_col_stat['group_max'] = groups['salary'].max()
    group_col_stat['group_min'] = groups['salary'].min()
    group_col_stat['group_std'] = groups['salary'].std()
    group_col_stat['group_median'] = groups['salary'].median()
    return group_col_stat
```

```python
cate_col = ['jobType', 'degree', 'major', 'industry']

group_col_stat = groupby_feature(encoded_df, cate_col)
```

```python
def merge_df(df1, df2, keys, fillna=False):
    df1 = pd.merge(df1, df2, on=keys, how='left')
    if fillna:
        df1.fillna(0, inplace=True)
    return df1

encoded_df = merge_df(encoded_df, group_col_stat, cate_col, fillna=True)
```

```python
encoded_df.columns
```

```python
feature_train = encoded_df.drop(['jobId', 'companyId', 'salary'], axis=1)
target_train = encoded_df['salary']
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA

def train_model(model, feature_df, target_df, num_procs, mean_mse, cv_std):
    neg_mse = cross_val_score(model, feature_df, target_df, cv=2, n_jobs=num_procs, scoring='neg_mean_squared_error')
    mean_mse[model] = -1.0*np.mean(neg_mse)
    cv_std[model] = np.std(neg_mse)
    
models = []
mean_mse = {}
cv_std = {}

num_procs = 4
verbose_lvl = 0

lr = LinearRegression()
lr_scaled_pca = make_pipeline(StandardScaler(), PCA(), LinearRegression())
rf = RandomForestRegressor(n_estimators=60, n_jobs=num_procs, max_depth=15, min_samples_split=80, max_features=8, verbose=verbose_lvl)                              
gbm = GradientBoostingRegressor(n_estimators=40, max_depth=7, loss='ls', verbose=verbose_lvl)

models.extend([lr, lr_scaled_pca, rf, gbm])
```

```python
def print_summary(model, mean_mse, cv_std):
    print('\nModel:\n', model)
    print('Average MSE:\n', mean_mse[model])
    print('Standard deviation during CV:\n', cv_std[model])
```

```python
from sklearn.model_selection import cross_val_score

for model in models:
    train_model(model, feature_train, target_train, num_procs, mean_mse, cv_std)
    print_summary(model, mean_mse, cv_std)
```

```python
best_model = min(mean_mse, key=mean_mse.get)
best_model
```
